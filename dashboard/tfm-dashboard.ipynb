# Dashboard

## 1. Librerias

```python
import time
import json
import boto3
import pandas as pd
import matplotlib.pyplot as plt
```

## 2. Configuración global

```python
BUCKET = "tfm-mu-bd"
PREFIX = "outputs/predictions/"

s3 = boto3.client("s3")

seen = set()
rows = []
daily = None
```

```python
BACKGROUND = "#ffffff"

LOW_COLOR = "#335c67"      
MEDIUM_COLOR = "#e09f3e"  
HIGH_COLOR = "#e63946"    

TEXT_COLOR = "#1d3557"
GRID_COLOR = "#457b9d"
CREAM_COLOR = "#f1faee"
NAVY_COLOR = "#0d1b2a"

RISK_COLORS = {
    "BAJO": LOW_COLOR,
    "MEDIO": MEDIUM_COLOR,
    "ALTO": HIGH_COLOR
}
```

## 3. Dashboard

```python
plt.style.use("default")
plt.ion()

fig, axes = plt.subplots(1, 3, figsize=(18, 5))
fig.patch.set_facecolor(BACKGROUND)

for ax in axes:
    ax.set_facecolor(BACKGROUND)
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)
    ax.spines["left"].set_color(TEXT_COLOR)
    ax.spines["bottom"].set_color(TEXT_COLOR)


def get_new_keys():
    response = s3.list_objects_v2(Bucket=BUCKET, Prefix=PREFIX)

    keys = []
    for obj in response.get("Contents", []):
        k = obj["Key"]
        if k not in seen:
            keys.append(k)
    return keys


def load_batch(keys):
    data = []
    for k in keys:
        obj = s3.get_object(Bucket=BUCKET, Key=k)
        data.append(json.loads(obj["Body"].read()))
    return data


while True:

    new_keys = get_new_keys()

    if new_keys:

        new_data = load_batch(new_keys)
        df_new = pd.DataFrame(new_data)

        rows.extend(new_data)
        seen.update(new_keys)

        df_new["date"] = pd.to_datetime(df_new["timestamp"]).dt.date

        partial = (
            df_new
            .groupby(["date", "risk"])
            .size()
            .unstack(fill_value=0)
        )

        daily = partial if daily is None else daily.add(partial, fill_value=0)

    if rows:

        df = pd.DataFrame(rows)
        risk_counts = df["risk"].value_counts()

        # Distribución de riesgo
        axes[0].clear()
        axes[0].set_facecolor(BACKGROUND)

        bar_colors = [
            RISK_COLORS.get(r.upper(), TEXT_COLOR)
            for r in risk_counts.index
        ]

        axes[0].bar(
            risk_counts.index,
            risk_counts.values,
            color=bar_colors
        )

        axes[0].set_title("Distribución de riesgo", color=TEXT_COLOR)
        axes[0].tick_params(colors=TEXT_COLOR)
        axes[0].grid(axis="y", alpha=0.25, color=GRID_COLOR)

        # Proporción de riesgo
        axes[1].clear()
        axes[1].set_facecolor(BACKGROUND)

        axes[1].pie(
            risk_counts.values,
            labels=risk_counts.index,
            colors=bar_colors,
            autopct="%1.1f%%"
        )

        axes[1].set_title("Proporción de riesgo", color=TEXT_COLOR)

        # Temporal
        axes[2].clear()
        axes[2].set_facecolor(BACKGROUND)

        if daily is not None:

            for col in daily.columns:

                axes[2].plot(
                    daily.index,
                    daily[col],
                    marker="o",
                    linewidth=2.5,
                    label=col
                )

        axes[2].set_title("Evolución por día", color=TEXT_COLOR)
        axes[2].tick_params(colors=TEXT_COLOR)
        axes[2].grid(alpha=0.25, color=GRID_COLOR)
        axes[2].legend(frameon=False)

    plt.tight_layout()
    plt.pause(2)
```
